De aqu√≠ en adelante, se utilizar√°n diferentes modelos de transformers para realizar las 3 tareas:
- Resumen del texto
- An√°lisis de sentimiento del texto
- Y traducci√≥n texto del espa√±ol al ingl√©s

Para ello, se utilizar√°n diferentes "pipelines" (de la plataforma de hugging face) para implementar diferentes modelos de transformer que sean capaces de realizar las 3 tareas diferentes.


### Resumen del texto

```python
# Ahora se aplica el pipeline con el modelo especializado en res√∫menes al texto de entrada con par√°metros espec√≠ficos
tranformerResumen = pipeline(
    "summarization", # Se pone en el pipeline que resuma el texto
    model="csebuetnlp/mT5_multilingual_XLSum", # Se pone el modelo "multilingue" especializado en res√∫menes
    device=-1  # Forzar uso de CPU en Colab para funcionar mejor
    )
```

```python
# Ahora se aplica el modelo al texto a resumir con sus parametros
textoResumido = tranformerResumen(
    textoDeEntrada, # "string" con el texto
    min_length = 50, # Longitud m√≠nima de 50 tokens (o palabras m√°s o menos)
    max_length = 200, # Longitud m√°xima de 200 tokens (o palabras m√°s o menos)
    do_sample = False # Desactiva la "aleatoriedad" para obtener un resultado m√°s estable
    )
```

```python
# Se imprime el resumen generado por el modelo
textoResumido[0]['summary_text']
```

Como se puede observar, ofrece un resultado satisfactorio, resumeniendo el texto con un m√°ximo de 200 palabras

### An√°lisis del sentimiento del texto

En este caso, se va a implementar el an√°lisis de sentimiento comparando 2 modelos y ver que resultados de los 2 proporciona y se observar√° cual de ellos funciona mejor


```python
# Ahora se implementa el pipeline con un modelo para el an√°lisis de sentimiento
textoSentimientoM1 = pipeline(
    "sentiment-analysis", # Se pone en el pipeline que analice el sentmiento del texto
    model="finiteautomata/beto-sentiment-analysis", # Se pone un modelo preentrenado que est√© especializado para el analisis de sentimiento en espa√±ol
    device=-1
    )
```

```python
# Se aplica el modelo de an√°lisis de sentmiento al texto completo
textoConSentimientoTodoTextoM1 = textoSentimientoM1(textoDeEntrada)
```

```python
# Se imprime el resultado del analisis de sentimiento del texto completo
textoConSentimientoTodoTextoM1
```

Como se puede observar, el resultado del modelo sale con la etiqueta de "negativo" y con porcentaje de confianza de 0.63, algo "impreciso". Esto puede ser debido a que el modelo es sensible a algunas "etiquetas", porque el texto en s√≠ no da indicios de ser "negativo".

```python
# Ahora se implementa el pipeline con otro modelo para el an√°lisis de sentimiento y comparar resultados
textoSentimientoM2 = pipeline(
    "sentiment-analysis", # Se pone en el pipeline que analice el sentmiento del texto
    model="tabularisai/multilingual-sentiment-analysis", # Se pone un modelo preentrenado que est√© especializado para el analisis de sentimiento en m√∫ltiples idiomas
    device=-1
    )
```

```python
# Se aplica el modelo de an√°lisis de sentmiento al texto completo
textoConSentimientoTodoTextoM2 = textoSentimientoM2(textoDeEntrada)
```

```python
# Se imprime el resultado del analisis de sentimiento del texto completo
textoConSentimientoTodoTextoM2
```

Y aqui como se puede observar, el resultado de este modelo es con etiqueta "neutral" y con un porcentaje de confianza de 0.87. Un resultado mejor y algo m√°s "entendible" que el del modelo anterior porque si se analiza el texto en s√≠ es m√°s descriptivo que negativo.

### Traducci√≥n al ingl√©s del texto

```python
# Ahora se implementa el pipeline con el modelo para traducir el texto al ingl√©s, para ello se utilizar√° un modelo especializado en traducci√≥n del texto del espa√±ol a ingl√©s
traducirTexto = pipeline(
    "translation", # Se pone en el pipeline que traduzca el texto
    model = "Helsinki-NLP/opus-mt-es-en", # Se pone un modelo preentrenado especializado en traducir los textos de espa√±ol a ingl√©s
    device=-1
    )
```

```python
# Se aplica el modelo de traducci√≥n al texto
textoTraducido = traducirTexto(textoDeEntrada)
```

```python
# Se imprime el texto traducido
textoTraducido[0]['translation_text']
```

Como se puede observar, traduce el texto del espa√±ol al ingl√©s

```python
# Ahora se compara ambos textos para ver si est√° bien la traducci√≥n
print("** TEXTO ORIGINAL **")
print(textoDeEntrada) # Se imprime el texto original
print("\n")
print("** TEXTO TRADUCIDO **")
textoTraducidoParaImprimir = textoTraducido[0]['translation_text']
# Se imprime la traduccion haciendo saltos de linea cada 120 caracteres con el m√©todo "fill" de textwrap
textoTraducidoParaImprimir = textwrap.fill(textoTraducidoParaImprimir, width=120)
print(f'\n{textoTraducidoParaImprimir}')
```

Como se puede observar, la traducci√≥n est√° bien hecha en su conjunto, solo tiene un peque√±o cambio en el final.
